## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
# Juntando os dataframes de investimento 1 e 2
investimentos_total = pd.concat([investimentos_1, investimentos_2])
display(investimentos_total)
print('-'*40)
# Arquivo de investimento total - quantidade de linhas
n_linhas_inv, n_cols_inv = investimentos_total.shape
display(f'A tabela de investimento total possui {n_linhas_inv} linhas e {n_cols_inv} colunas')
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
# Removendo linhas duplicadas do dataframe
investimentos_total_S_duplicados = investimentos_total.drop_duplicates()
display(investimentos_total_S_duplicados)
print('-'*40)
# Arquivo de investimento total sem linhas duplicadas - quantidade de linhas
n_linhas_inv_S_duplicados, n_cols_inv_S_duplicados = investimentos_total_S_duplicados.shape
display(f'A tabela de investimento total sem linhas duplicadas possui {n_linhas_inv_S_duplicados} linhas e {n_cols_inv_S_duplicados} colunas')
```

## 4 - Junte o dataframe de investimento com o de cadastro
Dica: use a chave id

```python
# Juntando os dataframes de investimento e de cadastro
df_total = cadastros.merge(investimentos_total_S_duplicados, how='left', on="id")
display(df_total)
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
# Ordenando o dataframe pelo id com sort_values e redefinindo o índice
df_total_sorted = df_total.sort_values('id', ascending=True).reset_index(drop=True)
display(df_total_sorted)
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
# Criando uma função para mostrar dados faltantes
def mostrarDadosFaltantes (dados):
    print("Coluna - Quantidade de dados faltantes - Porcentagem de dados faltantes")
    for i in dados.columns:
        print(i,"-", dados[i].isnull().sum(), f"- {round((dados[i].isnull().sum()/len(dados))*100, 2)}%")
```

```python
# Mostrando os dados faltantes do dataframe
mostrarDadosFaltantes(df_total_sorted)
```

```python
# Criando uma cópia do dataframe para tratar os dados faltantes
df_total_sorted_SemDadosFaltantes = df_total_sorted.copy()
```

```python
# Para as colunas de Ativos (Ativo01 até Ativo07), substituímos os valores de faltantes por 0 (zero). Desse modo mostraria o mesmo resultado, pois o investidor colocar R$0,00 é o mesmo de não colocar nenhum valor no ativo.
df_total_sorted_SemDadosFaltantes[["Ativo01","Ativo02", "Ativo03","Ativo04", "Ativo05","Ativo06", "Ativo07"]] = df_total_sorted_SemDadosFaltantes[["Ativo01","Ativo02", "Ativo03","Ativo04", "Ativo05","Ativo06", "Ativo07"]].fillna(value=0)

# Para as notas (Nota1 e Nota2), quando pelo menos uma das notas foi dada, adicionamos as notas que foram dadas na outra amostragem pelo mesmo participante. 
df_total_sorted_SemDadosFaltantes['Nota1'] = df_total_sorted_SemDadosFaltantes['Nota1'].fillna(value=df_total_sorted_SemDadosFaltantes['Nota2'].map(lambda x: x if pd.notnull(x) else None))
df_total_sorted_SemDadosFaltantes['Nota2'] = df_total_sorted_SemDadosFaltantes['Nota2'].fillna(value=df_total_sorted_SemDadosFaltantes['Nota1'].map(lambda x: x if pd.notnull(x) else None))

# Quando as duas notas eram faltantes, adicionamos a média de cada uma das notas. 
df_total_sorted_SemDadosFaltantes['Nota1'] = df_total_sorted_SemDadosFaltantes['Nota1'].fillna(value=df_total_sorted_SemDadosFaltantes['Nota1'].mean())
df_total_sorted_SemDadosFaltantes['Nota2'] = df_total_sorted_SemDadosFaltantes['Nota2'].fillna(value=df_total_sorted_SemDadosFaltantes['Nota2'].mean())

# Verificando se todos os dados faltantes foram preenchidos
mostrarDadosFaltantes(df_total_sorted_SemDadosFaltantes)
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
# Criar coluna "Total" somando o valor de todos os ativos por investidor
df_final['Total'] = df_final[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].sum(axis=1)
display(df_final.head())
```

```python
# Criar coluna "Media" com a média das notas 1 e 2 (duas casas decimais)
df_final['Media'] = df_final[['Nota1', 'Nota2']].mean(axis=1).round(2)
display(df_final.head())
```

```python
# Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
df_final['Dominio'] = df_final['Email'].apply(lambda email: email[(email.find('@'))+1:])
display(df_final.head())
```

```python
# Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
## Mudar o nome da coluna "Salário" para "Renda"
df_final = df_final.rename(columns={'Salario': 'Renda'})
## Tratar a coluna (sem caracter $ e tipo numérica)
df_final['Renda'] = df_final['Renda'].str[1:]
df_final['Renda'] = df_final['Renda'].astype(float)
display(df_final.head())
```

```python
# Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
## Verificando valores da coluna "Sexo"
print(df_final['Sexo'].unique())
## Mudando "Fem" para "Feminino" e "Masc" para "Masculino"
df_final['Sexo'] = df_final['Sexo'].map(lambda sexo: 'Feminino' if sexo == 'Fem' else 'Masculino')
display(df_final.head())
```

```python
# Transformar os dados da coluna Media em inteiro
df_final['Media'] = df_final['Media'].map(lambda x: int(x))
df_final['Media'].head()
```

```python
# Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50
def categorias(media):
    if media >= 80:
        return 'promotor'
    elif media >=50:
        return "neutro"
    else:
        return "detrator"
    
df_final['NPS'] = df_final['Media'].map(categorias)
display(df_final[["Media",'NPS']])
```
